```python
import matplotlib.pyplot as plt
import numpy as np

# Estados:
naranja=0
azul=1
verde=2
rojo=3

# Tomando que empezamos en naranja
naranja = 1
azul = 0
verde = 0
rojo = 0

n = 25
naranja_i = []
azul_i = []
verde_i = []
rojo_i = []

for i in range(n):
    nuevo_naranja = 0.25*azul + 0.5*verde + 0.25*rojo
    nuevo_azul = 0.25*naranja + 0.5*verde + 0.25*rojo
    nuevo_verde = 0.5*naranja + 0.25*azul + 0.25*rojo
    nuevo_rojo = 0.25*naranja + 0.5*azul + 0.25*verde

    naranja = nuevo_naranja
    azul = nuevo_azul
    verde = nuevo_verde
    rojo = nuevo_rojo

    naranja_i.append(naranja)
    azul_i.append(azul)
    verde_i.append(verde)
    rojo_i.append(rojo)

# Graficar resultados
plt.plot(naranja_i, label='Naranja', color='orange')
plt.plot(azul_i, label='Azul', color='blue')
plt.plot(verde_i, label='Verde', color='green')
plt.plot(rojo_i, label='Rojo', color='red')
plt.xlabel('Nmero de lanzamientos')
plt.ylabel('Probabilidad')
plt.title('Frecuencia de probabilidades en los sectores')
plt.legend()
plt.grid(True)
plt.show()

print("Probabilidades finales:")
print(f"Naranja: {naranja_i[-1]:.4f}")
print(f"Azul: {azul_i[-1]:.4f}")
print(f"Verde: {verde_i[-1]:.4f}")
print(f"Rojo: {rojo_i[-1]:.4f}")
```

Podemos notar que despus de varios lanzamientos es igualmente probable caer en cualquiera de los sectores.
